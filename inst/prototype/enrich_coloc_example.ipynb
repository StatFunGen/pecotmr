## 1. load pkgs

```r
library(susieR)
library(coloc)
library(pecotmr)
library(tidyverse)
library(data.table)
pkgs <- list.files("/mnt/vast/hpc/homes/rf2872/codes/pecotmr/R", full.names = TRUE)
for(i in pkgs){
    source(i)
}
```

## 2. enrichment 

```r
enrich_res <- xqtl_enrichment_wrapper(gwas_files = gwas_path,
                        xqtl_files = qtl_path,
                        gwas_finemapping_obj = NULL,
                        xqtl_finemapping_obj = c("Mic","susie_result_trimmed"),
                        gwas_varname_obj = NULL,              
                        xqtl_varname_obj = c("Mic","variant_names"))
```

## 3. coloc with prior

```r
coloc_res <- coloc_wrapper(
  xqtl_file = qtl_path, gwas_files = gwas_path,
    gwas_finemapping_obj = NULL, xqtl_finemapping_obj = c("Mic","susie_result_trimmed"),
    gwas_varname_obj = NULL, xqtl_varname_obj = c("Mic","variant_names"),
    LD_meta_file_path = "/mnt/vast/hpc/csg/data_public/20240120_ADSP_LD_matrix/ld_meta_file.tsv", 
    p1 = enrich_res$`Alternative (coloc) p1`, p2 = enrich_res$`Alternative (coloc) p2`,
    p12 = enrich_res$`Alternative (coloc) p12`
)
```
